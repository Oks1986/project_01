```python
def quarterof (month) :
    months = ['Январь',
         'Февраль',
         'Март',
         'Апрель',
         'Май',
         'Июнь',
         'Июль',
         'Август',
         'Сентябрь',
         'Октябрь',
         'Ноябрь',
         'Декабрь']
    q=(month-1)//3+1
    n=['первого','второго', 'третьего', 'четвертого']
    print('Месяц №', month, '('+months[month-1]+') является частью', n[q-1], 'квартала')
num =int(input('напишите номер месяца'))
quarterof(num)
```
